```python
import pandas as pd
import numpy as np
import re

df=pd.read_excel('D:/DataScience/Advent/Day4.xlsx', header=None,skip_blank_lines=False)

df_num=df.to_numpy()
Sep=np.argwhere(pd.isnull(df_num)); 
Sep1=Sep[:,0]+1
Sep2=Sep[:,0]

Sep1=np.insert(Sep1,0,0)
Sep2=np.append(Sep2,len(df_num))

df_list=list(df.values.flatten())

Num=len(Sep1)
df_person=list();

i=0;
for i in range(0,Num):
    combine_information=''
    j=Sep1[i];
    for j in range(Sep1[i],Sep2[i]):        
        combine_information=combine_information+df_list[j]+' '
    df_person.append(combine_information)
```

```python

Keys=['byr','iyr','eyr','hgt','hcl','ecl','pid']

i=0;num=0;id=[];
for i in range(0,len(df_person)):
    x=df_person[i]
    j=0
    Flag=100
    for j in range(0,len(Keys)):
        if x.find(Keys[j])==-1:
            flag=-1
        else:
            flag=1        
        Flag=Flag+flag;
    
    if Flag==107:
        num=num+1
        id.append(i)

print(num)
print(id)
```

```python
Dic=[]
for iper in id:
    x=df_person[iper]
    x=x.split(' ')
    list_key=[];
    list_value=[];
    for j in range (0,len(x)-1):
        y=x[j]   
        y=y.split(":")
        list_key.append(y[0])
        list_value.append(y[1])
    dic=dict(zip(list_key,list_value ) )
    Dic.append(dic)
    
df1=pd.DataFrame(Dic)
df1=df1.drop('cid',axis=1)
df1.head()
```

```python
df1['byr']=df1['byr'].apply(int)
df1['iyr']=df1['iyr'].apply(int)
df1['eyr']=df1['eyr'].apply(int)

df1=df1[df1.byr>=1920]
df1=df1[df1.byr<=2002]

df1=df1[df1.iyr>=2010]
df1=df1[df1.iyr<=2020]

df1=df1[df1.eyr>=2020]
df1=df1[df1.eyr<=2030]

filter_ecl=['amb','blu','brn','gry','grn','hzl','oth']
df1=df1[df1.ecl.isin(filter_ecl)]

df1=df1[df1.hcl.str.contains('^\\#')]
df1=df1[df1.hcl.str.len()==7]

df1=df1[df1.pid.str.len()==9]
df1['pid']=df1['pid'].apply(int)
df1=df1[df1.pid>=0]
df1=df1[df1.pid<=999999999]


df_cm=df1[df1.hgt.str.contains('cm')]
df_in=df1[df1.hgt.str.contains('in')]
df_cm['hgt'] = df_cm['hgt'].map(lambda x: str(x)[:-2])
df_in['hgt'] = df_in['hgt'].map(lambda x: str(x)[:-2])

df_cm.hgt=df_cm.hgt.apply(int)
df_in.hgt=df_in.hgt.apply(int)

df_cm=df_cm[df_cm.hgt>=150]
df_cm=df_cm[df_cm.hgt<=193]

df_in=df_in[df_in.hgt>=59]
df_in=df_in[df_in.hgt<=76]

len(df_cm)+len(df_in)
```
